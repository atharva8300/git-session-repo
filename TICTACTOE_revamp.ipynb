```python
from IPython.display import clear_output
import random


def display_board(board):
    clear_output()
    print(f"{board[7]}|{board[8]}|{board[9]}")
    print( "_._._")
    print(f"{board[4]}|{board[5]}|{board[6]} ")
    print( "_._._")
    print(f"{board[1]}|{board[2]}|{board[3]}")


def player_choice(player1,player2):
    while (not((player1 == 'x') or (player1 == 'o'))):
        player1 = input("player 1 choose your character : x/o")
    if player1 == 'x':
        return (('x','o'))
    else:
        return (('o','x'))

    
def whogoesfirst():
    return random.randint(1,2)


def checkplayer(board,player):
    x,y,r,l = 1,1,1,3
    win = 0
    while (x <= 7 and win==0): #check rows starting from 1,4 and 7
        if(board[x] == board[x+1] == board[x+2] == player):
            print(f"{player} won")
            win = 1 
        x += 3
    while (y<=3 and win ==0):
        if(board[y]==board[y+3]==board[y+6]==player):
            print(f"{player} won")
            win = 1 
        y += 1
    if(board[r]==board[r+4]==board[r+8]==player):
        print(f"{player} won")
        win = 1
    if(board[l]==board[l+2]==board[l+4]==player):
        print(f"{player} won")
        win = 1
    return win


def boardfull(board):
    for x in board:
        if x == ' ':
            return 0
    return 1


replay = 1
while(replay):
    board = [" "]*10
    board[0] = 'f'
    display_board(board)
    player1 = ''
    player2 = ''
    player1,player2 = player_choice(player1,player2)
    print(f"player1 = {player1} , player2 = {player2}")
    r = whogoesfirst()
    print(r)

    if r==1: #player 1 goes first
        print('player1 goes first')
        for f in range(0,100):
            if f%2 == 0:
                print("player 1's chance : ")
                choice = 0
                while(True):
                    choice = input("through numpad enter position : ")
                    if choice in "123456789":
                        break
                    else:
                        print("invalid input , enter number between 1-9")
                board[int(choice)] = player1
                display_board(board)
                if (checkplayer(board,player1)):
                    print('player 1 won')
                    break
                elif (boardfull(board)):
                    print('tie')
                    break
            else:
                print("player 2's chance : ")
                choice = 0
                while(True):
                    choice = input("through numpad enter position : ")
                    if choice in "123456789":
                        break
                    else:
                        print("invalid input , enter number between 1-9")
                board[int(choice)] = player2
                display_board(board)
                if (checkplayer(board,player2)):
                    print('player 2 won')
                    break
                elif (boardfull(board)):
                    print('tie')
                    break
                
    else: #player 2 goes first
        print('player 2 goes first')
        for f in range(0,100):
            if f%2 == 0:
                print("player 2's chance : ")
                choice = 0
                while(True):
                    choice = input("through numpad enter position : ")
                    if choice in "123456789":
                        break
                    else:
                        print("invalid input , enter number between 1-9")
                board[int(choice)] = player2
                display_board(board)
                if (checkplayer(board,player2)):
                    print('player 2 won')
                    break
                elif (boardfull(board)):
                    print('tie')
                    break
            else:
                print("player 1's chance : ")
                choice = 0
                while(True):
                    choice = input("through numpad enter position : ")
                    if choice in "123456789":
                        break
                    else:
                        print("invalid input , enter number between 1-9")
                board[int(choice)] = player1
                display_board(board)
                if (checkplayer(board,player1)):
                    print('player 1 won')
                    break
                elif (boardfull(board)):
                    print('tie')
                    break
    
    
    final = ''
    while (not(final == 'y' or final == 'n')):
        final = input('do you want to play again : y/n ?')
        if final == 'y':
            replay = 1
        else:
            replay = 0
```
